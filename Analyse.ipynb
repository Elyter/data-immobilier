```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv('data.csv')

df
```

```python
df = df[df['valeur_fonciere'].notnull() & df['surface_reelle_bati'].notnull()]
df
```

```python
df["prix_m2"] = df["valeur_fonciere"] / df["surface_reelle_bati"]

df["zscores"] = stats.zscore(df["prix_m2"])

df["zscores_valeur_fonciere"] = stats.zscore(df["valeur_fonciere"])

df

df["prix_moyen_commune"] 1
```

```python
df_filtered = df[(df['zscores'] > 3) | (df['zscores'] < -3)]
df_filtered['zscores']
df_filtered
```

```python
df = df[(df['zscores'] < 3) & (df['zscores'] > -3)]
df['zscores']
df
```

```python
df.to_sql('dvf', 'sqlite:///data.db', if_exists='replace', index=False)
```
